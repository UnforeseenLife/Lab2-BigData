```python
from pygeocoder import Geocoder
from time import sleep
```

```python
SOfile = open("SnowyOwl.csv", "r", errors="ignore");
AHfile = open("AnnasHummingbird.csv", "r", errors="ignore");
BSfile = open("BlackSwift.csv", "r", errors="ignore");
Bfile = open("Brambling.csv", "r", errors="ignore");
HDfile = open("HarlequinDuck.csv", "r", errors="ignore");
LBfile = open("LazuliBunting.csv", "r", errors="ignore");



SOlist = [];
AHlist = [];
BSlist = [];
Blist = [];
HDlist = [];
LBlist = [];
SOlocations = [];
AHlocations = [];
BSlocations = [];
Blocations = [];
HDlocations = [];
LBlocations = [];



for a in SOfile:
    SOlist.append(a);
del SOlist[0];

for b in AHfile:
    AHlist.append(b);
del AHlist[0];

for c in BSfile:
    BSlist.append(c);
del BSlist[0];

for d in Bfile:
    Blist.append(d);
del Blist[0];

for e in HDfile:
    HDlist.append(e);
del HDlist[0];

for f in LBfile:
    LBlist.append(f);
del LBlist[0];



for g in range(len(SOlist)):
    SOlist[g] = SOlist[g].split(",");

for h in range(len(AHlist)):
    AHlist[h] = AHlist[h].split(",");

for i in range(len(BSlist)):
    BSlist[i] = BSlist[i].split(",");

for j in range(len(Blist)):
    Blist[j] = Blist[j].split(",");

for k in range(len(HDlist)):
    HDlist[k] = HDlist[k].split(",");

for l in range(len(LBlist)):
    LBlist[l] = LBlist[l].split(",");

SOWfile = open("SOneighborhoods.txt", "w");
AHWfile = open("AHneighborhoods.txt", "w");
BSWfile = open("BSneighborhoods.txt", "w");
BWfile = open("Bneighborhoods.txt", "w");
HDWfile = open("HDneighborhoods.txt", "w");
LBWfile = open("LBneighborhoods.txt", "w");
    

#Following code influneced by Chris Albon, http://chrisalbon.com/python/geocoding_and_reverse_geocoding.html
for SOlocate in range(len(SOlist)):
    SOresults = Geocoder.reverse_geocode(float(SOlist[SOlocate][20]), float(SOlist[SOlocate][21]));
    SOlocations.append(SOresults.neighborhood); 
        #can use route for street names (more specific)
        #use city for city
        #use county for county (more specific than city)
        #use neighborhood for neighborhood (more specific than route)
    SOWfile.write(SOresults.neighborhood);
    sleep(0.1); #Google API limits to 10 pings per second
print(SOlocations);

"""
for AHlocate in range(len(AHlist)):
    AHresults = Geocoder.reverse_geocode(float(AHlist[AHlocate][20]), float(AHlist[AHlocate][21]));
    AHlocations.append(AHresults.neighborhood); 
    AHWfile.write(AHresults.neighborhood);
    sleep(0.1); #Google API limits to 10 pings per second
print(AHlocations)
"""
print();

for BSlocate in range(len(BSlist)):
    BSresults = Geocoder.reverse_geocode(float(BSlist[BSlocate][20]), float(BSlist[BSlocate][21]));
    BSlocations.append(BSresults.neighborhood); 
    BSWfile.write(BSresults.neighborhood);
    sleep(0.1); #Google API limits to 10 pings per second
print(BSlocations);

for Blocate in range(len(Blist)):
    Bresults = Geocoder.reverse_geocode(float(Blist[Blocate][20]), float(Blist[Blocate][21]));
    Blocations.append(Bresults.neighborhood);
    BWfile.write(Bresults.neighborhood);
    sleep(0.1); #Google API limits to 10 pings per second
print(Blocations);

for HDlocate in range(len(HDlist)):
    HDresults = Geocoder.reverse_geocode(float(HDlist[HDlocate][20]), float(HDlist[HDlocate][21]));
    HDlocations.append(HDresults.neighborhood); 
    HDWfile.write(HDresults.neighborhood);
    sleep(0.1); #Google API limits to 10 pings per second
print(HDlocations);

for LBlocate in range(len(LBlist)):
    LBresults = Geocoder.reverse_geocode(float(LBlist[LBlocate][20]), float(LBlist[LBlocate][21]));
    LBlocations.append(LBresults.neighborhood);
    LBWfile.write(LBresults.neighborhood);
    sleep(0.1); #Google API limits to 10 pings per second
print(LBlocations);



#FOR API WITH NO USAGE LIMIT https://www.fcc.gov/general/census-block-conversions-api

"""print(SOlist);
print();
print(AHlist);
print();
print(BSlist);
print();
print(Blist);
print();
print(HDlist);
print();
print(LBlist);"""
print();
```
